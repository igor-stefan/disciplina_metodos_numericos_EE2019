```python
import numpy as np
import matplotlib.pyplot as plt

def criar_vetor_x(a,h,n):
  x=np.zeros(n+1,dtype=float)
  x[0]=a
  for i in range(n):
    x[i+1]=x[i]+h
  return x  

def trapezio_simples(f,a,b):
  return (b-a)*(f(a)+f(b))/2.0

def trapezio_composta(f,a,b,n):
  h=(b-a)/n
  soma=0.0
  x=criar_vetor_x(a,h,n)
  for i in range(1,n):
    soma+=(2*f(x[i]))
  return (h/2)*(f(x[0])+f(x[-1])+soma)

def simpson_3_simples(f,a,b):
  h1=(b-a)/2
  n1=2
  soma=0.0
  x=criar_vetor_x(a,h1,n1)
  for i in range(n1+1):
    if(np.remainder(i,2)):
      soma+=(4*f(x[i]))
    else:
      soma+=f(x[i])
  return h1*soma/3

def simpson_3_composta(f,a,b,n):
  h=(b-a)/n
  soma=0.0
  x=criar_vetor_x(a,h,n)
  for i in range(1,n):
    if(np.remainder(i,2)):
      soma+=(4*f(x[i]))
    else:
      soma+=(2*f(x[i]))
  soma+=f(x[0])
  soma+=f(x[-1])
  return h*soma/3


def calc_erro(x1,x2):
  return np.abs((x1-x2)/x1)*100
```
