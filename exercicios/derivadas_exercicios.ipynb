1. Calcular $f'(x)$ para $x=3$ utilizando $h = 0.1, 0.05, 0.025$ através usando diferença finita para frente. Calcule o erro real, sabendo que o resultado analítico é $f'(3) = 270.05$

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

VR=270.05

def derivada_p(f,x,h):
  return (f(x+h)-f(x))/h;

def derivada_r(f,x,h):
  return (f(x)-f(x-h))/h;

def derivada_c(f,x,h):
  return (f(x+h) - f(x-h))/(2*h);

def g(x):
  return 2*np.exp(1.5*x);

def derivada_f(f,x,h):
  return (f(x+h)-f(x))/h

def calcular_erro(x1,x2):
  return np.abs((x1-x2)/x1)*100

print ('a =',derivada_f(g,3,.1))
print('erro =',(derivada_f(g,3,.1)-VR)*100/derivada_f(g,3,.1))
print('\nb =',derivada_f(g,3,.05))
print('erro =',(derivada_f(g,3,.05)-VR)*100/derivada_f(g,3,.05))
print('\nc =',derivada_f(g,3,.025))
print('erro =',(derivada_f(g,3,.025)-VR)*100/derivada_f(g,3,.0125))
print('\nd =',derivada_f(g,3,.0125))
print('erro =',(derivada_f(g,3,.0125)-VR)*100/derivada_f(g,3,.0125))
```

7. A Lei de Faraday caracteriza a queda de tensão em um indutor como:

$$V_L = L \frac{d_i}{d_t}$$

onde $V_L = $ queda de tensão, $L$ = indutância (em Henrys), $i = corrente (A)$, e $t = tempo$. Determine a queda de tensão para cada valor de tempo a partir dos dados abaixo e para um indutância de 4H

t|i
--|--
0|0
0.1|0.16
0.2|0.32
0.3|0.56
0.5|0.84
0.7|2.0

```python
def far(x):
  if(x==0.0):
    return 0.0;
  elif(x==.1):
    return .16;
  elif(x==.2):
    return .32;
  elif(x==.3):
    return .56;
  elif(x==.5):
    return .84;
  elif(x==.7):
    return 2.0;

def derivada_pr(f,x,h):
  arg=round(x+h,5)
  return (f(arg)-f(x))/h;

def derivada_rr(f,x,h):
  arg=round(x-h,5)
  return (f(x)-f(arg))/h;

def derivada_cr(f,x,h):
  args=round(x+h,5)
  argd=round(x-h,5)
  return (f(args) - f(argd))/(2*h);


print('vl(0) =',4*derivada_pr(far,0.0,0.1))
print('\nvl(0.1) =',4*derivada_cr(far,0.1,.1))
print('\nvl(0.2) =',4*derivada_cr(far,0.2,.1))
print('\nvl(0.3) =',4*derivada_cr(far,0.3,.2))
print('\nvl(0.5) =',4*derivada_cr(far,0.5,.2))
print('\nvl(0.7) =',4*derivada_rr(far,0.7,.2))

```

8. A taxa de resfriamento de um corpo pode ser expressada como:

$$\frac{dT}{dt} = -k(T - T_a)$$

onde $T = $ temperatura do corpo (ºC), $T_a = $ temperatura do meio (ºC), e $k=$ constante de proporcionalidade (por minuto). Logo, essa equação especifica que a taxa de resfriamento é proporcional a diferença de temperaturas de um corpo e do meio que o envolve. Se uma bola de metal aquecida até a temperatura de 80ºC é colocada dentro de um balde com água a uma temperatura constante de $T_a$ = 20ºC, a temperatura da bola muda de acordo com a tabela abaixo:

Tempo, min | T ºC
--|--
0|80
5|44.5
10|30.0
15|24.1
20|21.7
25|20.7

Utilizando diferenciação numérica determine $dT/dt$ para cada valor de tempo. Trace o gráfico $dT/dt$ versus $T - T_a$ e empregue regressão linear para encontrar $k$

```python
x8=np.array([0.0,5.0,10.0,15.0,20.0,25.0],dtype=float)
y8=np.array([80.0,44.5,30.0,24.1,21.7,20.7],dtype=float)
dif_t=np.zeros(x8.shape[0]) #diferenca de temperatura

for i in range(x8.shape[0]):
  dif_t[i]=y8[i]-20

def t(x):
  for i in range(x8.shape[0]):
    if x==x8[i]:
      return y8[i]

dT=np.zeros(x8.shape[0],dtype=float) #taxa de resfriamento

for k in range(x8.shape[0]):
  if k==0:
    dT[k]=derivada_p(t,x8[k],5)
  elif k < x8.shape[0]-1:
    dT[k]=derivada_c(t,x8[k],5)
  else:
    dT[k]=derivada_r(t,x8[k],5)


plt.plot(x8,dT,color='green') #temperatura e taxa de resfriamento
plt.plot(x8,dif_t,color='red') #temperatura e diferenca de temperatura
plt.grid(1)
plt.show()


dif_t=dif_t.reshape((-1, 1)) #transformado para matriz coluna

reg=LinearRegression().fit(dif_t,dT) #regressão diferenca de temperatura e taxa de resfriamento
k=reg.coef_ #coeficiente
print('k =',k)
```

```python
plt.plot(x8,dT,color='red') #grafico temperatura e taxa de resfriamento
plt.plot(x8,k*dif_t,color='black') #grafico temperatura e k*diferenca de temperatura
plt.grid(1)
plt.show()
```

10. Estime $f'(1)$ e $f''(1)$ o mais preciso possível a partir dos dados:


x|$f(x)
--|--
0.97|0.85040
1.00|0.84147
1.05|0.82612

Justifique a escolha do método.

```python
def newton(x,val_x,val_y):
  coeficientes = np.zeros(val_x.shape[0])
  m_val=np.zeros((val_x.shape[0],val_x.shape[0]),dtype=float)
  m_controle=np.zeros((val_x.shape[0],val_x.shape[0]),dtype=bool)
  res=0;
  diferenca=1;
  for i in range (coeficientes.shape[0]):
    coeficientes[i]=pd(0,i,m_val,m_controle,val_x,val_y)
    res+=diferenca*coeficientes[i]
    diferenca*=x-val_x[i]
  #print('Valores dos coeficientes = ',coeficientes);
  #print('f(',x,') = ', res,sep='')
  return res;

def pd(inicio,fim,m_val,m_controle,val_x,val_y):
  if m_controle[inicio,fim]:
    return m_val[inicio,fim] #se ja tiver sido calculado, retorna o valor referente
  if(inicio==fim):
    return val_y[inicio] 
  m_val[inicio,fim]=(pd(inicio+1,fim,m_val,m_controle,val_x,val_y)-pd(inicio,fim-1,m_val,m_controle,val_x,val_y))/(val_x[fim]-val_x[inicio])
  m_controle[inicio,fim]=1
  return m_val[inicio,fim];

x10=np.array([0.97,1.0,1.05],dtype=float)
y10=np.array([0.85040,0.84147,0.82612],dtype=float)
newx=np.zeros(10,dtype=float)
newx[0]=0.97
newy=np.zeros(10,dtype=float)
for k in range(9):                    #montando novo vetor de valores de x e de y calculados com interpolação
  newx[k+1]=newx[k]+.01
for q in range(newx.shape[0]):
  newy[q]=newton(newx[q],x10,y10)

def r(x):
  for j in range(newx.shape[0]):
    if x==newx[j]:
      return newy[j]

f_1=np.zeros(10,dtype=float)

for i in range(newx.shape[0]-1):
  if i==0:
    f_1[i]=derivada_pr(r,newx[i],0.03)
  elif i==3 or i==6:
    f_1[i]=derivada_cr(r,newx[i],0.03)
  elif i==8:
    f_1[i]=derivada_rr(r,newx[i],0.03)

# for j in range(f_1.shape[0]):  
#   if f_1[j]!=0.0:
#     print('f_1[',j,']=',f_1[j])


def n1(x):
  for k in range(newx.shape[0]):
    if x==newx[k]:
      return f_1[k] #vetor com valores da derivada
 
print("f'(1) = ",derivada_cr(r,1.0,0.03))
print("f''(1) = ",derivada_cr(n1,1.0,0.03))
#primeiro foi realizada uma interpolação para descobrir valores que seriam uteis e permitiriam calcular a derivada central com o intervalo de 0.03
#em seguida foram calculadas as primeiras derivadas e armazenadas num vetor
#por ultimo, foi calculada a segunda derivada com base na primeira derivada
#foi utilizada a derivada central pois é mais precisa
```
