```python
import numpy as np
np.set_printoptions(formatter={'float': lambda x: "{:.4f}".format(x)}) # configura a impressão dos floats

def eliminacao(matriz):
  for i in range(matriz.shape[0]):
    pivo=matriz[i,i] #pivo elemento da diagonal principal
    equacao_pivo=matriz[i,:] #linha que representa a equacao pivo
    j=i+1 #para acessar as linhas a frente da pivo
    while(j<matriz.shape[0]): #enquanto nao chegar a ultima linha
      eliminador=matriz[j,i]/pivo #calculo o elemento que fara a eliminacao do termo
      matriz[j,:]=matriz[j,:]-(eliminador*equacao_pivo) #atualiza a linha que vai ter o elemento eliminado
      j=j+1 #acessa a proxima linha para fazer nova eliminacao
  return matriz 

def substituicao(matriz):
  solucoes=np.zeros(matriz.shape[0],dtype=float) #vetor de solucoes
  solucoes[-1]=matriz[-1,-1]/matriz[-1,-2] #a primeira solucao é conhecida
  soma=0
  for i in range(-(matriz.shape[0]-2),1): #vai da ultima linha ate a primeira. de n a 0
    for j in range(matriz.shape[0]): #para calcular as somas possiveis
      if(j!=-i): #epossivel somar quando eu conheco o valor de x descoberto na linha anterior. acesso o vetor de solucoes atraves do indice j
        soma+=solucoes[j]*matriz[-i,j] #faco as somas possiveis
    solucoes[-i]=(matriz[-i,-1]-soma)/matriz[-i,-i] #calculo a solucao
    soma=0 #zero o valor de soma para o proximo calculo na linha acima   
  print(solucoes)
  for k in range (matriz.shape[0]):
    print('x',k+1,' = ',solucoes[k],sep='') 
  return solucoes

def gauss_ingenuo(matriz_sistema):
  return substituicao(eliminacao(matriz_sistema))


def pivotamento(matriz):
  for i in range(matriz.shape[0]):
    maior_abs=np.abs(matriz[i,i])
    maior=matriz[i,i]
    for q in range(i+1,matriz.shape[0]): #for para procurar o maior numero nas linhas do pivo
      if(np.abs(matriz[q,i])>maior_abs): #se o numero encontrado for maior que o inicial
        maior=matriz[q,i] #entao o maior numero é o encontrado
        matriz[[i,q]]=matriz[[q,i]] #troco as linhas da matriz
    #print('i=',i,'\nmatriz=',matriz)
    pivo=maior #pivo elemento da diagonal principal
    equacao_pivo=matriz[i,:] #linha que representa a equacao pivo
    j=i+1 #para acessar as linhas a frente da pivo
    while(j<matriz.shape[0]): #enquanto nao chegar a ultima linha
      eliminador=matriz[j,i]/pivo #calculo o elemento que fara a eliminacao do termo
      matriz[j,:]=matriz[j,:]-(eliminador*equacao_pivo) #atualiza a linha que vai ter o elemento eliminado
      j=j+1 #acessa a proxima linha para fazer nova eliminacao
  return matriz 

def substituicao(matriz):
  solucoes=np.zeros(matriz.shape[0],dtype=float) #vetor de solucoes
  solucoes[-1]=matriz[-1,-1]/matriz[-1,-2] #a primeira solucao é conhecida
  soma=0
  for i in range(-(matriz.shape[0]-2),1): #vai da ultima linha ate a primeira. de n a 0
    for j in range(matriz.shape[0]): #para calcular as somas possiveis
      if(j!=-i): #e possivel somar quando eu conheco o valor de x descoberto na linha anterior. acesso o vetor de solucoes atraves do indice j
        soma+=solucoes[j]*matriz[-i,j] #faco as somas possiveis
    solucoes[-i]=(matriz[-i,-1]-soma)/matriz[-i,-i] #calculo a solucao
    soma=0 #zero o valor de soma para o proximo calculo na linha acima   
  print(solucoes)
  for k in range (matriz.shape[0]):
    print('x',k+1,' = ',solucoes[k],sep='') 
  return solucoes

def gauss_pivotamento(matriz_sistema):
  return substituicao(pivotamento(matriz_sistema))


def verificar_dominancia(matriz):
  for t in range(matriz.shape[0]):
    soma=0.0
    for j in range(matriz.shape[1]):
      if t!=j:
        soma+=matriz[t,j]
    if(np.abs(matriz[t,t])<np.abs(soma)):
      print('O sistema não possui convergencia garantida')
      return False
  print('O sistema converge')
  return True


def erro(novo,ant,taxa,vetor):
  a=np.zeros(novo.shape[0])
  for i in range(novo.shape[0]):
    a[i]=np.abs((novo[i]-ant[i])*100/novo[i])
    vetor.append(a[i])
  #print('erro=',a) #possibilita printar o vetor mostrando todos os erros encontrados, respectivamente na ordem para x1,x2,...,xn
  if max(a)<taxa: #se o erro for menor do que o requisitado na funcao, true para parar o laço
    return True
  return False

def gauss_seidel(matriz,iteracoes,t_erro,relaxamento,lambd):
  #relaxamento acelera convergencia se lambda>1 e o sistema convergir
  # tipicamente usado para fazer com que um sistema não-convergente passe a convergir se lambda<1
  #lambd entre 0 e 2
  atual=np.zeros(matriz.shape[0],dtype=float) #vetor com as soluçoes da iteração
  anteriores=atual.copy() #vetor para soluçoes anteriores para calculo do erro
  v_erro=[] #vetor para acumular os valores do erro
  soma=0.0 #variavel auxiliar temporaria
  for k in range(iteracoes): #ate o numero max de iterações
    for i in range(matriz.shape[0]): #em cada linha
      soma=0.0 #zera a variavel
      for j in range(matriz.shape[0]): #em cada coluna da linha
        if j!=i: #se nao esta na diagonal principal
          soma+=matriz[i,j]*atual[j] #soma an's*xn's
      atual[i]=(matriz[i,-1]-soma)/matriz[i,i] #calcula o valor de xn
      if relaxamento: #se usa a tecnica do relaxamento
        atual[i]=atual[i]*lambd+(1-lambd)*anteriores[i] #valor atual
    if (k>0): #apos a primeira iteração
      if(erro(atual,anteriores,t_erro,v_erro)): #calculo o erro
        break #se esta menor que o pedido breca o laço
      anteriores=atual.copy() #atualiza os valores do vetor anterior para as proximas iteraçoes
  print('numero de iteracoes: ',k+1) 
  #print('penultima iteração:')
  #for k in range (matriz.shape[0]):
    #print('x',k+1,' = ',anteriores[k],sep='')
  #print('resultado final:')  
  for k in range (matriz.shape[0]):
    print('x',k+1,' = ',atual[k],sep='') 
  return atual 

def dlu(matriz_sistema):
  matriz_sistema=matriz_sistema[matriz_sistema[:,0].argsort(kind='quicksort')][::-1] #ordena o sistema na ordem decrescente pela primeira coluna
  #print('matriz_sistema=\n',matriz_sistema)
  u=matriz_sistema[:,:-1] #matriz dos coeficientes
  b=matriz_sistema[:,-1] #matriz dos resultados
  l=np.identity(u.shape[0],dtype=float) #identidade, matriz l
  l,u,b=pivotamento_lu(u,l,b) #chama função para encontrar l e u
  return l,u,b

def pivotamento_lu(matriz_u,matriz_l,b):
  p=np.identity(matriz_u.shape[0])
  for i in range(matriz_u.shape[0]-1):
    maior_abs=np.abs(matriz_u[i,i]) #maior em modulo
    maior=matriz_u[i,i] #maior = primeiro numero da coluna
    for q in range(i+1,matriz_u.shape[0]): #for para procurar o maior numero nas linhas do pivo
      if(np.abs(matriz_u[q,i])>maior_abs): #se o numero encontrado for maior que o inicial
        maior=matriz_u[q,i] #entao o maior numero é o encontrado
        matriz_u[[i,q]]=matriz_u[[q,i]] #troco as linhas da matriz
        p[[i,q]]=p[[q,i]] #troco as linhas na matriz identidade
    pivo=maior 
    equacao_pivo=matriz_u[i,:] #linha que representa a equacao pivo
    j=i+1 #para acessar as linhas a frente da pivo
    while(j<matriz_u.shape[0]): #enquanto nao chegar a ultima linha
      eliminador=matriz_u[j,i]/pivo #calculo o elemento que fara a eliminacao do termo
      matriz_u[j,:]=matriz_u[j,:]-(eliminador*equacao_pivo) #atualiza a linha que vai ter o elemento eliminado
      matriz_l[j,i]=eliminador #preenche a matriz l
      j=j+1 #acessa a proxima linha para fazer nova eliminacao
  b=np.matmul(p,b) #multiplica b pela matriz identidade p
  return matriz_l,matriz_u,b

def matriz_d(l,b):
  d=np.zeros(l.shape[0],dtype=float) #numero de itens que compoem o vetor d
  soma=0.0
  for i in range(l.shape[0]): #para encontrar os valores de d, faz-se a substituiçao progressiva
    soma=0.0
    for j in range(l.shape[0]):
      if (j!=i):
        soma+=l[i,j]*d[j]
    d[i]=b[i]-soma
  return d

def solucao(u,d):
  solucoes=np.zeros(u.shape[0],dtype=float) #vetor de solucoes
  solucoes[-1]=d[-1]/u[-1,-1] #a primeira solucao é conhecida
  soma=0.0
  for i in range(-(u.shape[0]-2),1): #vai da ultima linha ate a primeira. de n a 0
    for j in range(u.shape[0]): #para calcular as somas possiveis
      if(j!=-i): #e possivel somar quando eu conheco o valor de x descoberto na linha anterior. acesso o vetor de solucoes atraves do indice j
        soma+=solucoes[j]*u[-i,j] #faco as somas possiveis
    solucoes[-i]=(d[-i]-soma)/u[-i,-i] #calculo a solucao
    soma=0.0 #zero o valor de soma para o proximo calculo na linha acima   
  print(solucoes)
  for k in range (u.shape[0]):
    print('x',k+1,' = ',solucoes[k],sep='') 
  return solucoes

def slu(matriz):
  l1,u1,b1=dlu(matriz)
  d1=matriz_d(l1,b1)
  return solucao(u1,d1)

def solucao_inversa(u,d):
  solucoes=np.zeros(u.shape[0],dtype=float) #vetor de solucoes
  solucoes[-1]=d[-1]/u[-1,-1] #a primeira solucao é conhecida
  soma=0.0
  for i in range(-(u.shape[0]-2),1): #vai da ultima linha ate a primeira. de n a 0
    for j in range(u.shape[0]): #para calcular as somas possiveis
      if(j!=-i): #e possivel somar quando eu conheco o valor de x descoberto na linha anterior. acesso o vetor de solucoes atraves do indice j
        soma+=solucoes[j]*u[-i,j] #faco as somas possiveis
    solucoes[-i]=(d[-i]-soma)/u[-i,-i] #calculo a solucao
    soma=0.0 #zero o valor de soma para o proximo calculo na linha acima    
  return solucoes

def pivotamento_lu_inversa(matriz_u,matriz_l):
  for i in range(matriz_u.shape[0]-1):
    maior=matriz_u[i,i] #maior = primeiro numero da coluna
    pivo=maior 
    equacao_pivo=matriz_u[i,:] #linha que representa a equacao pivo
    j=i+1 #para acessar as linhas a frente da pivo
    while(j<matriz_u.shape[0]): #enquanto nao chegar a ultima linha
      eliminador=matriz_u[j,i]/pivo #calculo o elemento que fara a eliminacao do termo
      matriz_u[j,:]=matriz_u[j,:]-(eliminador*equacao_pivo) #atualiza a linha que vai ter o elemento eliminado
      matriz_l[j,i]=eliminador #preenche a matriz l
      j=j+1 #acessa a proxima linha para fazer nova eliminacao
  return matriz_l,matriz_u

def dlu_inversa(matriz_sistema):
  u=matriz_sistema[:,:-1] #matriz dos coeficientes
  b=matriz_sistema[:,-1] #matriz dos resultados
  l=np.identity(u.shape[0],dtype=float) #identidade, matriz l
  l,u=pivotamento_lu_inversa(u,l) #chama função para encontrar l e u
  return l,u

def calcular_inversa(matriz):
  a=np.zeros(matriz.shape[0],dtype=float)
  matriz=np.vstack((matriz,a))
  a=np.zeros(matriz.shape[0],dtype=float)
  a=np.reshape(a,(-1,1))
  matriz=np.hstack((matriz,a))
  
  l,u=dlu_inversa(matriz)
  l1=l[:-1,:-1]
  u1=u[:-1,:]
  res=np.zeros((l1.shape[0],l1.shape[0]),dtype=float)
  b1=np.identity(l1.shape[0])
  for i in range(l1.shape[0]):
    d1=matriz_d(l1,b1[:,i])
    res[:,i]=solucao_inversa(u1,d1)
  return res

def calcular_erro(x1,x2):
  return np.abs((x1-x2)/x1)*100

def verificar_solucao(solutions,sistema,tol):
  res=0
  for i in range(sistema.shape[0]):
    soma=0.0
    for j in range(sistema.shape[1]-1):
      soma+=(round(solutions[j],10)*round(sistema[i,j],10))
    if(np.abs(soma-sistema[i,-1])<=tol):
      res+=1
    else:
      break
  if(res==sistema.shape[0]):
    print ('Solucões corretas!')
  else:
    print('Solucões incorretas! :/')
```

## Modelagem matemática

Pela lei de *Kirchhoff*, a soma das correntes que passam em cada nó do circuito é nula. Pela lei de *Ohm* a corrente elétrica que flui do nó *j* para o nó *k* de um circuito é 

$$I_{jk} = \frac{V_j-V_k}{R_{jk}}$$

sendo $V_j$ e $V_k$ as tensões nos nós $j$  e $k$, respectivamente, e $R_{jk}$ a resistência no arco _jk_. A corrente $I$ é expressa em ampères e a resistência $R$, em ohms. As duas leis combinadas permitem o cálculo da tensão em cada nó do circuito.

No nó 1, pela lei de _Kirchoff_,

$$I_{A1} + I_{21} + I_{31} + I_{41} = 0 $$

Utilizando a lei de _Ohm_, 

$$ \frac{0-V_1}{1} + \frac{V_2-V_1}{1} + \frac{V_3-V_1}{2} + \frac{V_4-V_1}{2} = 0 $$

simplificando:

$$ -6V_1 + 2V_2 + V_3 + V_4 = 0$$


Determinando as equações dos nós 2, 3 e 4, constrói-se um sistema de equações lineares. O vetor solução fornece a tensão em cada nó do circuito elétrico.

Construa o resto do sistema e utilizando dois métodos de resolução numérica encontre as tensões em cada um dos nós.

```python
a8=np.array([[3,0,1,0,127.0],[3,-4,1,0,0.0],[1,2,0,-3,0],[9,2,-11,6,0]])
print('pivotamento:')
gauss_pivotamento(a8)
print('\ndecomposiçao lu:')
x_8=slu(a8)
print('\ngauss ingenuo:')
gauss_ingenuo(a8)
print('\ngauss seidel:')
gauss_seidel(a8,100,.01,0,0.2)
print('\n')
verificar_solucao(x_8,a8,.8)
```
