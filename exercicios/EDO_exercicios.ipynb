```python
import numpy as np
import matplotlib.pyplot as plt

#################################################################### RK4 #############################################################################
def k1(f,x,y):
  return f(x,y)

def k2(f,h,x,y,k1):
  return f(x+(h/2), y+(h*k1/2))

def k3(f,h,x,y,k2):
  return f(x+(h/2), y+(h*k2/2))

def k4(f,h,x,y,k3):
  return f(x+h, y+h*k3)

def rk4(f,h,x,y):
  k_1=k1(f,x,y)
  k_2=k2(f,h,x,y,k_1)
  k_3=k3(f,h,x,y,k_2)
  k_4=k4(f,h,x,y,k_3)
  return y+((h/6)*(k_1+2*k_2+2*k_3+k_4))

# o metodo consiste em inciar a interação com os valores de x0 e y0 fornecidos
# a variavel h indica o passo ou intervalo entre um y conhecido e um y a ser calculado
# quando a interação inicia, o h passa a ser o x e o valor retornado pela função rk4 é o novo y

#################################################################### EULER ############################################################################

def euler(f,h,x,y):
  return y+(h*f(x,y))

#################################################################### RK2 ##############################################################################

def heun(f,h,x,y):
  k_1=f(x,y)
  k_2=f((x+h),y+(k_1*h))
  return y+((k_1/2)+(k_2/2))*h


def ponto_medio(f,h,x,y):
  k_1=f(x,y)
  k_2=f(x+(h/2),y+(k_1*h/2))
  return y+(k_2*h)


def ralston(f,h,x,y):
  k_1=f(x,y)
  k_2=f(x+(3*h/4),y+(3*k_1*h/4))
  return y+(((k_1/3)+(2*k_2/3))*h)
```

1. Dada a seguinte equação diferencial:

$$\frac{dy}{dx} + 0.4y(x) = 3e^{-x}$$

$$\frac{dy}{dt} = yt^3 - 1.5y$$

    2.1 Utilizando o método de Euler com h = 0.5 e 0.25
    2.2 Utilizando o método do ponto médio com h = 0.5
    2.3 Utilizando RK4 com h = 0.5

```python
#2.1.1 EULER
#DEFININDO INTERVALO E A FUNÇÃO
x0=0
y0=1
h2=.5
h3=.25
def q(x,y):
  return (y*(x**3)-(1.5*y))

###############################################################################################################

val_x=np.zeros(5, dtype=float) 
val_y=np.zeros(5, dtype=float)
val_y[0]=y0

for i in range(4):
  val_x[i+1]=val_x[i]+h2 #o primeiro valor novo de x é x0+h; ou seja, o inicio mais o passo
  val_y[i+1]=euler(q,h2,x0,y0) #o primeiro valor novo de y é o resultado da funcao euler
  x0=val_x[i+1]
  y0=val_y[i+1]

print('Valores x = ',val_x)
print('Valores y = ',val_y)
print('Y(2) = ',val_y[i+1])
plt.plot(val_x,val_y,'ro-')
plt.grid(1)
plt.show()
```


3. Resolva o PVI no intervalo de x = 0 a 1 utilizando h = 0.25. Mostre todos os resultados em um mesmo gráfico.

$$\frac{dy}{dx} = (1+4x)\sqrt{y}$$

    3.1 Utilizando o método de Euler
    3.2 Utilizando o método de Heun
    3.3 Utilizando o método de Ralston
    3.4 Utilizando RK4

```python
h3=.25
def q3(x,y):
  return (1+(4*x))*np.sqrt(y)
#########################################################################  
#3.1 EULER
valores_x=np.zeros(5,dtype=float);
valores_y=np.zeros(5,dtype=float);
valores_y[0]=2.0;
valores_x[0]=0.0
for i in range(4):
  valores_x[i+1]=valores_x[i]+h3
  valores_y[i+1]=euler(q3,h3,valores_x[i],valores_y[i])

print('Valores x = ',valores_x)
print('Valores y = ',valores_y)
print('Y(1) = ',valores_y[i+1])
plt.plot(valores_x,valores_y,'mo-')
plt.grid(1)
plt.show()
```

4.  Resolva o seguinte problema de valor inicial no intervalo de x = -1 a 1 em que y(-1) = 1/e.
$$\frac{dy}{dx} = e^{2x+1}+3x^2+y$$

    4.1. Usando um método de segunda ordem com h=0.01
    
    4.2 Usando RK4 com h=0.05

```python
def f41(x,y):
  return np.exp((2*x)+1)+(3*x**2)+(y)

h41=.01
#################################################################################
#4.1 METODO DE HEUN 
valores_x=np.zeros(201,dtype=float);
valores_y=np.zeros(201,dtype=float);
valores_y[0]=1/np.exp(1)
valores_x[0]=-1.0
for i in range(200):
  valores_x[i+1]=valores_x[i]+h41
  valores_y[i+1]=heun(f41,h41,valores_x[i],valores_y[i])

#print('Valores x = ',valores_x)
print('Valores y = ',valores_y)
print('Y(1) = ',valores_y[i+1])
plt.plot(valores_x,valores_y,'-')
plt.grid(1)
plt.show()
```

5. Dado o problema de valor inicial abaixo:
$$y'(x) = \frac{1}{x}(2y+x+1)$$
$$y(1) = 0.5$$

Calcule y(100) usando o método de Heun com h=0.2

```python
h5=.2
def q5(x,y):
  return (1/x)*((2*y)+x+1)
##########################################################
#5 RK2 HEUN COM H=.2

x5=np.zeros(496,dtype=float);
y5=np.zeros(496,dtype=float);
x5[0]=1.0
y5[0]=.5

for i in range(495):
  x5[i+1]=x5[i]+h5
  y5[i+1]=heun(q5,h5,x5[i],y5[i])

#print('Valores x = ',x5)
print('Valores y = ',y5)
print('Y(100) = ',y5[i+1])
plt.plot(x5,y5,'m-')
plt.grid(1)
plt.show()
```
