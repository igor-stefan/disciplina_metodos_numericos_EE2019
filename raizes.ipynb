```
#importa as bibliotecas
import numpy as np
import matplotlib.pyplot as plt
#funcao:
def f(x):
  return x**3-20;
```

#Função para exibir o gráfico
<p align="justify"> Abaixo está definida uma função para exibir qualquer gráfico. Como exemplo está a chamada para a função $f(x)=sin(x)$.
Obs: Atualizar a função no bloco acima.</p>

```
def plotar_grafico(x1,x2,i,f):
  intervalo=[] #cria um vetor para intervalo
  y=[] #cria um vetor para receber os valores de f(x)
  intervalo = np.arange(x1,x2,i); #associa o intervalo ao vetor
  for j in intervalo: #itera j no intervalo
    y.append(f(j)); #adiciona em y os valores calculados de f(j)
  plt.plot(intervalo,y); #plota o grafico x,y
  plt.grid(True); 
  plt.show();
  
  

plotar_grafico(-5.0,5.0,0.001,f);
```

#Método da Bisseção
<p align="justify"> Abaixo está definida uma função para encontrar as raízes, seguindo o método da bisseção tendo como entrada o inicio e o fim do intervalo, o número máximo de iterações, o erro e uma função $f(x)$. Como exemplo está a chamada para a função $f(x) = x^{3}-20 $. </p>

```
def bissec(x1,x2,max_it,eps,f):
  xm=[];
  it=0;
  erro=[];
  if (f(x1)*f(x2) > 0):
    print("Não é possível calcular as raízes com os valores de x1 e x2 dados");
    return;
  for i in range(max_it):
    it=it+1;
      xm.append((x1+x2)/2)
      if(f(x1)*f(xm[i]) < 0):
        x2 = xm[i];
      else:
        x1 = xm[i];
    if i != 0:
      erro.append((np.abs((xm[i]-xm[i-1])/xm[i]))*100)
      if(erro[-1]<eps):
        break;
  print(xm[-1])
  print("Número de iterações necessárias: {}".format(it))
  plt.plot(erro)
  plt.grid(True)
  plt.show()
  
bissec(1,4,50,0.005,f);
```

#Método das Cordas
<p align="justify"> Abaixo está definida uma função para encontrar as raízes seguindo o método das cordas tendo como entrada o inicio e o fim do intervalo, o número máximo de iterações, o erro e uma função $f(x)$. Como exemplo está a chamada para a função $f(x) = x^{3}-20 $. </p>

```
def cordas(x1,x2,max_it,eps,f):
  xm=[];
  it=0;
  erro=[];
  if (f(x1)*f(x2) > 0):
    print("Não é possível calcular as raízes com os valores de x1 e x2 dados");
    return;
  for i in range(max_it):
    it=it+1;
      xm.append(x2+(f(x2)/(f(x1)-f(x2)))*(x2-x1))
      if(f(x1)*f(xm[i]) < 0):
        x2 = xm[i];
      else:
        x1 = xm[i];
    if i != 0:
      erro.append((np.abs((xm[i]-xm[i-1])/xm[i]))*100)
      if(erro[-1]<eps):
        break;
  print(xm[-1])
  print("Número de iterações necessárias: {}".format(it))
  plt.plot(xm)
  plt.grid(True)
  plt.show()
  
cordas(0,5.4,50,0.005,f);
```

#Método de Newton

<p align="justify"> Abaixo está definida uma função para encontrar as raízes seguindo o método de Newton tendo como entrada o chute inicial, o número máximo de iterações, o erro e uma função $f(x)$. Como exemplo está a chamada para a função $f(x) = x^{3}-20 $. </p>

```
def derivada(ponto):
  return 3*(ponto**2);

def newton(x1,max_it,eps,f):
  it=0;
  xm=[x1];
  erro=[x1];
  for j in range(max_it):
    it=it+1;
    xm.append(xm[-1]-(f(xm[-1])/derivada(xm[-1])));
    erro.append((np.abs((xm[j]-xm[j-1])/xm[j]))*100);
    if(erro[-1]<eps):
      break;
  print(xm[-1])
  print("Número de iterações necessárias: {}".format(it))
  plt.plot(xm) #colocar plot(erro) para gráfico do erro
  plt.grid(True)
  plt.show()

newton(5,50,0.0005,f)
```

#Método da Secante

<p align="justify"> Abaixo está definida uma função para encontrar as raízes seguindo o método da secante tendo como entrada dois chutes iniciais, o número máximo de iterações, o erro e uma função $f(x)$. Como exemplo está a chamada para a função $f(x) = x^{3}-20 $. </p>

```
def secante(x1,x2,max_it,eps,f):
  it=0;
  xm=[x1,x2];
  erro=[x1,x2];
  for j in range(max_it):
    it=it+1
    xm.append(xm[j+1]-((f(xm[j+1])*(xm[j+1]-xm[j]))/(f(xm[j+1])-f(xm[j]))));
    erro.append((np.abs((xm[j]-xm[j-1])/xm[j]))*100);
    if(erro[-1]<eps):
      break;
  print(xm[-1])
  print("Número de iterações necessárias: {}".format(it))
  plt.plot(xm) #colocar plot(erro) para gráfico do erro
  plt.grid(True)
  plt.show()
  
secante(1.0,4.0,50,0.0005,f);
```
